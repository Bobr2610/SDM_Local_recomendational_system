```python
import pandas as pd
import numpy as np
```

```python
path = 'raw/train_wide.csv'

df = pd.read_csv(path)

df.head()
```

```python
df.drop(columns=['fecha_dato'], inplace=True)

df.head()
```

```python
df['sex'].isna().sum()
```

```python
df.dropna(subset=['sex'], inplace=True)
```

```python
df['sex'].isna().sum()
```

```python
df['age'].isna().sum()
```

```python
df.loc[df["age"] < 0, "age"] = np.nan
```

```python
df['age'].isna().sum()
```

```python
df.loc[(df["age"] < 14) | (df["age"] > 95), "age"] = np.nan
```

```python
df['age'].isna().sum()
```

```python
df.dropna(subset=['age'], inplace=True)
```

```python
df['age'].isna().sum()
```

```python
df['is_new_customer'].isna().sum()
```

```python
df['is_new_customer'].value_counts()
```

```python
df['seniority_months'].isna().sum()
```

```python
df.loc[df['seniority_months'] < 0, 'seniority_months'] = np.nan
```

```python
df['seniority_months'].isna().sum()
```

```python
df.dropna(subset=['seniority_months'], inplace=True)
```

```python
df['seniority_months'].isna().sum()
```

```python
df['region_code'].isna().sum()
```

```python
df['region_name'].isna().sum()
```

```python
df['segment'].isna().sum()
```

```python
df['segment'] = df['segment'].fillna('UNKNOWN')
```

```python
df['segment'].isna().sum()
```

```python
output_path = 'processed/v1/train_wide_cleaned.csv'

df.to_csv(output_path, index=False)
```
